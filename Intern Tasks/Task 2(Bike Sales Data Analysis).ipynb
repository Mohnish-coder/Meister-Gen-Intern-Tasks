```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel("/content/uncleaned bike sales data.xlsx")
data.info() # day, age_group, product_description, order_quntity are not filled
```

```python
data['Day']=data['Day'].fillna(5)
data.info()
```

```python
data['Age_Group'].unique()
```

```python
print(data[data['Age_Group'] == 'Adults (35-64)'].count())
print(data[data['Age_Group'] == 'Young Adults (25-34)'].count())
print(data[data['Age_Group'] == 'Youth (<25)'].count())
print(data[data['Age_Group'] == 'nan'].count())
```

```python
data['Age_Group']=data['Age_Group'].fillna("Adults (35-64)")
data.info()
```

```python
data['Product_Description'].value_counts()
```

```python
data['Product_Description']=data['Product_Description'].fillna("Mountain-200 Black, 46 ")
data.info()
```

```python
data['Order_Quantity'].value_counts()
```

```python
data['Order_Quantity']=data['Order_Quantity'].fillna(1)
data.isnull().count()
```

```python
data.info()
```

```python
data.boxplot()

```

```python
#revenue
q1 = np.percentile(data['Revenue'], 25)
q3 = np.percentile(data['Revenue'], 75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
revenue_data = data[(data['Revenue'] >= lower_bound) & (data['Revenue'] <= upper_bound)]
```

```python
#cost
q1 = np.percentile(data[' Cost '], 25)
q3 = np.percentile(data[' Cost '], 75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
cost_data = data[(data[' Cost '] >= lower_bound) & (data[' Cost '] <= upper_bound)]
```

```python
#profit
q1 = np.percentile(data[' Profit '], 25)
q3 = np.percentile(data[' Profit '], 75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
profit_data = data[(data[' Profit '] >= lower_bound) & (data[' Profit '] <= upper_bound)]
```

```python
#unit price
q1 = np.percentile(data[' Unit_Price '], 25)
q3 = np.percentile(data[' Unit_Price '], 75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
unitprice_data = data[(data[' Unit_Price '] >= lower_bound) & (data[' Unit_Price '] <= upper_bound)]
```

```python
# histogram
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(data[' Profit '], kde=True)
plt.title('Original Profit Distribution')

plt.subplot(1, 2, 2)
sns.histplot(profit_data[' Profit '],kde=True)
plt.title('Filtered Profit Distribution')

plt.show()      #The purpose of this code is to visually compare the distribution of the 'Profit' column before and after removing outliers. The left subplot represents the original distribution, and the right subplot represents the distribution after applying outlier filtering. This visual comparison allows you to observe the impact of outlier removal on the overall shape and characteristics of the 'Profit' distribution.
```

```python
profit_data_numeric = profit_data.apply(pd.to_numeric, errors='coerce')
sns.heatmap(data=profit_data_numeric, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap')
plt.show()
```
