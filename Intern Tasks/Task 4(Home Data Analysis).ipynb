```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/content/home_data.csv")
```

```python
data.isnull().sum()  #there is no null values
```

```python
#maximum price for each type of bedroom
bedroom_five=data[data['bedrooms']==5]
print("five bedroom type->",bedroom_five['price'].max())
bedroom_four=data[data['bedrooms']==4]
print('four bedroom type->',bedroom_four['price'].max())
bedroom_three=data[data['bedrooms']==3]
print('three bedroom type->',bedroom_three['price'].max())
bedroom_two=data[data['bedrooms']==2]
print('two bedroom type->',bedroom_two['price'].max())
```

```python
data['floors'].drop_duplicates()  #types of floors
```

```python
print('five rooms->\n',bedroom_five['floors'].value_counts())
print('four rooms->\n',bedroom_four['floors'].value_counts())
print('three rooms->\n',bedroom_three['floors'].value_counts())
print('two rooms->\n',bedroom_two['floors'].value_counts())
#count of floor types
```

```python
data['price'].min
```

```python
data['price'].max()
```

```python
# Plotting pie chart for Distribution of Homes Based on Number of Bedrooms
room_type_counts = data['bedrooms'].value_counts()
plt.pie(room_type_counts, labels=room_type_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral', 'lightgreen', 'orange'])
plt.title('Distribution of Homes Based on Number of Bedrooms')
```

```python
# Histogram of Prices
plt.figure(figsize=(10, 6))
plt.hist(data['price'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Home Prices')
plt.xlabel('Frequency')
plt.ylabel('Price')
plt.show()
#we can infere that high price homes availability are less in frequency(less in number) && low price homes availability are high in frequency(more in number)
```

```python
# Correlation Heatmap
plt.figure(figsize=(12, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Scatter Plot of Living Area vs. Price
plt.figure(figsize=(10, 6))
plt.scatter(data['sqft_living'], data['price'], alpha=0.5, color='green')
plt.title('Living Area vs. Price')
plt.xlabel('Living Area (sqft)')
plt.ylabel('Price')
plt.show()
```

```python
# Boxplots for Bedrooms and Bathrooms
plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
sns.boxplot(x='bedrooms', y='price', data=data, palette='pastel')
plt.title('Boxplot of Price for Each Bedroom Count')

plt.subplot(2, 1, 2)
sns.boxplot(x='bathrooms', y='price', data=data, palette='pastel')
plt.title('Boxplot of Price for Each Bathroom Count')

plt.tight_layout()
plt.show()
```

```python
# Assuming 'date' column is in datetime format
data['date'] = pd.to_datetime(data['date'])

# Time Series Plot for Prices Over Time
plt.figure(figsize=(12, 6))
plt.plot(data['date'], data['price'], marker='o', linestyle='-', color='purple')
plt.title('Time Series Plot of Home Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
